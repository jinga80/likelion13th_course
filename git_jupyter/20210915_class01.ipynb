```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#np.linspace(시작,끝, 데이터 개수)
x = np.linspace(0,14,100)
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4
```

```python
plt.plot(x,y1,x,y2,x,y3)
```

```python
plt.plot(x,y1,'o')
plt.plot(x,y2,'o')
plt.plot(x,y3,'o')
```

```python
x = np.linspace(0,14,100)
y1 = np.sin(x)
x = np.linspace(0,14,100)
y2 = np.cos(x)
x = np.linspace(0,14,100)
y3 = np.tan(x)
```

```python
plt.plot(x,y1,x,y2,x,y3)
```
