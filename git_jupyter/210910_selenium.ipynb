```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
import time

url = 'https://pythonstart.github.io/web/'
start = time.time()
chrome_path = r'/opt/homebrew/bin/chromedriver'
driver = webdriver.Chrome(executable_path=chrome_path)
driver.get(url)
```

```python
selected_id = driver.find_element_by_id('rank')
print(selected_id)
print(selected_id.tag_name)# 해당 연결의 tag이름 
print(selected_id.text) # 해당 태그의 text정보
```

```python
selected_tag_p = driver.find_element_by_tag_name('h1')
print(selected_tag_p)
print(selected_tag_p.tag_name)
print(selected_tag_p.text)
```

```python
a_tag = driver.find_element_by_id('rank')
print(a_tag.text)
```

```python
a_tag = driver.find_elements_by_name('link_get')
print(a_tag)
```

```python
for one in a_tag:
    print(one.text)
```
